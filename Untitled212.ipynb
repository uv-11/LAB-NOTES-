```python
def dfs(st, goal, graph, vis=None):
    if vis is None:
        vis = set()

    print(st)

    if st == goal:
        return True

    vis.add(st)

    for child in graph[st]:
        if child not in vis:
            if dfs(child, goal, graph, vis):
                return True

    return False
```

```python
graph= {
    'A': ['B','C'],
    'B': ['A','E','D','F'],
    'C': ['A','D','F'],
    'D': ['B','C','E','F'],
    'E': ['B','D'],
    'F': ['B','C','D']
}
st='A'
goal='F'
```

```python
dfs('A','F',graph)
```

```python
from collections import deque

def bfs_path(graph, start, goal):
    queue = deque([[start]])
    visited = {start}

    while queue:
        path = queue.popleft()
        node = path[-1]

        if node == goal:
            return path

        for neighbour in graph[node]:
            if neighbour not in visited:
                visited.add(neighbour)
                new_path = path + [neighbour]
                queue.append(new_path)

    return None
```

```python
bfs_path(graph,'A','F')
```
