```python
import networkx as nx

graph = nx.Graph()

while True:
    ch = int(input("enter choice (1 to enter edge and 0 to exit): "))
    if ch == 1:
        n1 = input("enter first node: ").strip().upper()
        n2 = input("enter second node: ").strip().upper()
        wt = int(input("enter weight: "))
        graph.add_edge(n1, n2, weight=wt)
    else:
        break

print("\ngraph connections")
for node, neighbour in graph.adj.items():
    print(node, "-->", {n: d["weight"] for n, d in neighbour.items()})

start = input("\nenter start node: ").strip().upper()
goal = input("enter goal node: ").strip().upper()

heuristic_values = {}
heuristic_values[goal] = 0

print("\nEnter heuristic values")
for node in graph.nodes():
    if node != goal:
        h = int(input(f"enter heuristic value for {node}: "))
        heuristic_values[node] = h

def get_heuristic(u, v):
    return heuristic_values[u]

path = nx.astar_path(graph, start, goal, heuristic=get_heuristic, weight="weight")
total_cost = nx.astar_path_length(graph, start, goal, heuristic=get_heuristic, weight="weight")

print("\nShortest path:", " -> ".join(path))
print("Total cost:", total_cost)
```
