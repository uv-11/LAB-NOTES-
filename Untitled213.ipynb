```python
def dfs(st, goal, graph, vis=None):
    if vis is None:
        vis = set()

    print(st)

    if st == goal:
        return True

    vis.add(st)

    for child in graph[st]:
        if child not in vis:
            if dfs(child, goal, graph, vis):
                return True

    return False
```

```python
graph= {
    'A': ['B','C','D'],
    'B': ['A','E'],
    'C': ['A','F'],
    'D': ['A','G'],
    'E': ['G','H'],
    'F': ['C','H'],
    'G': ['D','H'],
    'H': ['E','F','G']
}
dfs('A','H',graph,vis=None)
```

```python
from collections import deque

def bfs_path(graph, start, goal):
    queue = deque([[start]])
    print("deque",queue)
    visited = {start}

    while queue:
        path = queue.popleft()
        print
        node = path[-1]

        if node == goal:
            return path

        for neighbour in graph[node]:
            if neighbour not in visited:
                visited.add(neighbour)
                new_path = path + [neighbour]
                queue.append(new_path)

    return None
```
