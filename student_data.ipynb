```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

marks =[ 50,70,40]
print("list of marks - ", marks)
arr = np.array(marks)
print("Array of marks",arr)
student= {
    "id": [101,102,103],
    "name": [ "abc"  , "def" , "ghi" ],
    "Age": [18,19,17]
}
student["marks"]= arr
df= pd.DataFrame(student)
df
```

```python
plt.bar(df["name"],df["marks"])
plt.xlabel("Students")
plt.ylabel("Marks")
plt.title("students vs marks ")
plt.show()
```

```python

plt.plot(df["name"],df["marks"])
plt.xlabel("Students")
plt.ylabel("Marks")
plt.title("students vs marks ")
plt.show()
```

```python
plt.barh(df["name"],df["marks"])
plt.xlabel("Marks")
plt.ylabel("Students")
plt.title("students vs marks ")
plt.show()
```
